```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
import folium
import seaborn as sns

df = pd.read_pickle('data/REC_database.pkl')

pd.set_option('display.max_rows', None)
```

```python
#replace "na" with actual na value in system_size_kW
df['system_size_kW'].replace('na', 0, inplace=True)
df.isna().sum()
```

```python
df.memory_usage(index=True, deep=False)
```

```python
# number
df[['Start_Serial']]=df[['Start_Serial']].astype('int32')
df[['End_Serial']]=df[['End_Serial']].astype('int32')
df[['Range_ID']]=df[['Range_ID']].astype('int32')
df[['Public_Registered_Person_ID']]=df[['Public_Registered_Person_ID']].astype('int32')
df[['number_of_certificates']]=df[['number_of_certificates']].astype('int16')

# float 
df[['system_size_kW']]=df[['system_size_kW']].astype('float16')
df['system_size_kW'] = df['system_size_kW'].round(1)

# Date
df['Creation_Date_simple'] = pd.to_datetime(df['Creation_Date_simple'], format='%Y-%m-%d')

# Categorical
df['Owner' ]=df['Owner' ].astype('category')
df['Fuel_Source_Display_Name']=df['Fuel_Source_Display_Name'].astype('category')
df['Status.1']=df['Status.1'].astype('category')
df['State']=df['State'].astype('category')
df['Owner_Name']=df['Owner_Name'].astype('category')
df['Created_By']=df['Created_By'].astype('category')

# Categorical - ordinal
df['Dominent zone'] = pd.Categorical(df['Dominent zone'], ordered=True)
df['Generation_Year'] = pd.Categorical(df['Generation_Year'], ordered=True)

# bool   
df['Fuel_Source_Active']=df['Fuel_Source_Active'].astype('bool')         

# string
df['Accreditation_Code'] = df['Accreditation_Code'].astype('string')
```

```python
# Check categories
print('Owner:',df['Owner'].cat.categories, '-',df['Owner'].nunique(),'\n')
print('Fuel_Source_Display_Name:',df['Fuel_Source_Display_Name'].cat.categories, '-', df['Fuel_Source_Display_Name'].nunique(),'\n')
print('Status.1:',df['Status.1'].cat.categories, '-', df['Status.1'].nunique(),'\n')
print('State:',df['State'].cat.categories, '-', df['State'].nunique(),'\n')
print('Owner_Name:',df['Owner_Name'].cat.categories, '-', df['Owner_Name'].nunique(),'\n')
print('Created_By:',df['Created_By'].cat.categories, '-', df['Created_By'].nunique(),'\n')
print('Dominent zone:',df['Dominent zone'].cat.categories, '-', df['Dominent zone'].nunique(),'\n')
print('Generation_Year:',df['Generation_Year'].cat.categories, '-', df['Generation_Year'].nunique(),'\n')
```

```python
# remove these columns
df.drop(['Status'], axis=1, inplace=True) # only one category so not useful
df.drop(['Fuel_Source_Type' ], axis=1, inplace=True) # duplicate of Fuel_Source_Display_Name
df.drop(['Creation_Month' ], axis=1, inplace=True) # captured in Creation_Date_simple so duplicate info
df.drop(['Creation_Year' ], axis=1, inplace=True) # captured in Creation_Date_simple so duplicate info
df.drop(['Creation_Date' ], axis=1, inplace=True) # captured in Creation_Date_simple so duplicate info
```
